```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def ODE_RK2 (f, t0, tf, y0, n = 100) :

    t = np.linspace(t0, tf, n)
    y = list([y0])

    for i in range(n-1) :
        h = t[i+1] - t[i]
        k1 = h*f(t[i], y[-1])
        k2 = h*f(t[i]+(h/2.0), y[-1]+(k1/2.0))
        y.append(y[-1]+k2)

    y = np.array(y)

    return y
```

```python
def ODE_RK4 (f, t0, tf, y0, n = 100) :

    t = np.linspace(t0, tf, n)
    y = list([y0])

    for i in range(n-1) :
        h = t[i+1] - t[i]
        k1 = h*f(t[i], y[-1])
        k2 = h*f(t[i]+(h/2.0), y[-1]+(k1/2.0))
        k3 = h*f(t[i]+(h/2.0), y[-1]+(k2/2.0))
        k4 = h*f(t[i]+h, y[-1]+k3)
        y.append(y[-1]+(k1+2*k2+2*k3+k4)/6)

    y = np.array(y)

    return y
```

```python
def f(t, y) :
    return 4-3*y
```

```python
def F(t) :
    return (-1.0/3)*np.exp(-3*t)+(4.0/3)
```

```python
t = np.linspace(0, 2, 100)
yt = F(t)
```

```python
y_RK2 = ODE_RK2(lambda x, y :f(x, y), 0, 2, 1, 100)
y_RK4 = ODE_RK4(lambda x, y :f(x, y), 0, 2, 1, 100)
```

```python
plt.plot(t, yt, t, y_RK2, 'r')
```

```python
plt.plot(t, abs(yt-y_RK2))
```

```python
plt.plot(t, yt, t, y_RK4, 'r')
```

```python
plt.plot(t, abs(yt-y_RK4))
```

```python
plt.plot(t, abs(yt-y_RK2), 'y', t, abs(yt-y_RK4), 'r')
```
