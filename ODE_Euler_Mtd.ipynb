```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def ode_Euler(f, t0, tf, y0 = 0, n =100) :

    t = np.linspace(t0, tf, n)
    y = list([y0])
    for i in range(n-1):
        h = t[i+1] - t[i]
        y.append(y[-1]+h*f(t[i], y[-1]))

    y = np.array(y)

    return y 
```

```python
def f(t, y) :    #ODE
    return 4-3*y 
```

```python
def F(t) :
    return -1.0/3*np.exp(-3*t) + 4.0/3
```

```python
y = ode_Euler(lambda x, y: f(x, y), 0, 2, 1, 100)
```

```python
t = np.linspace(0, 2, 100)
yt = F(t)
```

```python
plt.plot(t, y, t, yt, 'r')
```

```python
plt.plot(t, abs(yt-y), 'r')
```
